```python
promesas = ((futbol['potential'] >= 82) & (futbol['age'] <= 21) & (futbol['value_eur'] <= 1800000) & (futbol["overall"] >=70))
promesas_id = list(futbol.loc[promesas]['player_id'])
top_promesas = futbol[futbol['player_id'].isin(promesas_id)].sort_values(by = ['player_id', 'fifa_version']).copy()
top_promesas
```

```python
# Compute the mean wage per FIFA version for both datasets
promesas_avg_wage = top_promesas.groupby("fifa_version")["wage_eur"].mean()
futbol_avg_wage = futbol.groupby("fifa_version")["wage_eur"].mean()

# Merge into a DataFrame for comparison
comparison_df = pd.DataFrame({"Top Promesas": promesas_avg_wage, "Futbol General": futbol_avg_wage})

print(comparison_df)
```

```python
# Filtrar jugadores por posici칩n y puntuaci칩n general
df_filtered = futbol[futbol["overall"] >90]

# Comparar jugadores por edad y puntuaci칩n
plt.figure(figsize=(10,6))
sns.scatterplot(x=df_filtered["age"], y=df_filtered["overall"], hue=df_filtered["club_name"])
plt.xlabel("Edad")
plt.ylabel("Puntuaci칩n General")
plt.title("Comparaci칩n de jugadores en FIFA")
plt.legend(title="Club")
plt.show()
```
