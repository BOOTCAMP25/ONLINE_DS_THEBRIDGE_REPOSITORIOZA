```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

# Train model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Compute MAE
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Obtener las importancias de las características
feature_importance = model.feature_importances_

# Ordenar características por importancia
indices = np.argsort(feature_importance)[::-1]
feature_names = X.columns

plt.figure(figsize=(10, 6))
plt.bar(range(X.shape[1]), feature_importance[indices], align="center")
plt.xticks(range(X.shape[1]), np.array(feature_names)[indices], rotation=90)
plt.xlabel("Característica")
plt.ylabel("Importancia")
plt.title("Importancia de Características en Random Forest")
plt.show()
```

🔹 ¿Por qué?

El MAE en entrenamiento es bajo (0.1677), lo que sugiere que el modelo aprende bien en los datos de entrenamiento.

🔧 ¿Cómo mejorarlo? Aquí tienes algunas estrategias para reducir el sobreajuste y mejorar la generalización:

✅ Reducir la complejidad del modelo

Baja max_depth en el Gradient Boosting para evitar árboles demasiado profundos.

Reduce n_estimators para evitar sobreaprendizaje en el conjunto de entrenamiento.

Usa learning_rate más bajo para suavizar el aprendizaje.

✅ Aumentar la cantidad de datos

Si tienes acceso a más datos, intenta expandir el conjunto de entrenamiento.

Usa técnicas de data augmentation si aplicable.

✅ Aplicar regularización

Usa subsample=0.8 en GradientBoostingRegressor para entrenar con muestras parciales y reducir el riesgo de sobreajuste.

En otros modelos como Ridge Regression, prueba aumentar el hiperparámetro alpha.

✅ Validación cruzada

Usa cross_val_score() para verificar si el modelo es estable en diferentes particiones de datos.

El MAE promedio en validación cruzada es bastante cercano al MAE de prueba (0.4376), lo que sugiere que el modelo tiene una generalización razonable, pero aún podría mejorarse.

🔹 ¿Qué significa esto?

Dado que la validación cruzada refleja el rendimiento en distintas particiones de los datos, el valor 0.4508 indica que el modelo mantiene un rendimiento estable.

Sin embargo, sigue siendo algo más alto que el MAE en entrenamiento (0.1677), lo que implica que el modelo aún tiene cierto nivel de sobreajuste.

🔧 Mejoras posibles: ✅ Reducir la complejidad del modelo:

Baja max_depth o n_estimators en el modelo de Gradient Boosting.

Usa subsample=0.8 para reducir el sobreajuste. ✅ Regularización:

Ajusta la learning_rate a un valor menor para hacer el aprendizaje más gradual.
